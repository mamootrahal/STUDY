# Регулярные выражения в Python и pandas

**Регулярные выражения** - это шаблоны, используемые для сопоставления последовательностей символов в строках.

Как найти в тексте все числа из четырех цифр или все email-адреса или все номера телефонов?

На веб-сайтах:

имеются репозитории готовых регулярных выражений.

````python
# импортируем модуль для работы с регулярными выражениями: https://docs.python.org/3/library/re.html
import re
````

Регулярные выражения обычно содержат различные специальные символы, которые называются метасимволами:

`[] {} () \ * + ^ $ ? . |`

Ниже перечислены некоторые предопределенные символьные классы и группы символов, с которыми они совпадают.

С метасимвола `\` начинается каждый из предварительно определенных *символьных классов*, каждый из которых совпадает с символом из конкретного набора.

- `\d` Любая цифра `(0–9)`
- `\D` Любой символ, кроме цифр
- `\s` Любой символ-пропуск (пробелы, табуляции, новые строки)
- `\S` Любой символ, кроме пропусков
- `\w` Любой символ слова (также называемый алфавитно-цифровым символом) — то есть любая буква верхнего или нижнего регистра, любая цифра или символ подчеркивания
- `\W` Любой символ, кроме символов слов

Чтобы любой метасимвол совпадал со своим литеральным значением, поставьте перед ним символ `\` (обратный слеш). Например, `\\` совпадает с обратным слешем `( \ )`, а `\$` совпадает со знаком `$`.


````python
# Проверим, что ширина экрана содержит 4 цифры:

if re.fullmatch(r'\d{4}', '1920'):
    print('yes')
else:
    print('no')

if re.fullmatch(r'\d{4}', '768	'):
    print('yes')
else:
    print('no')
````

В регулярном выражении `\d{4}` `\d` является символьным классом, представляющим цифру `(0–9)`.

*Символьный класс* — служебная последовательность в регулярном выражении, совпадающая с одним символом. Чтобы совпадение могло состоять из нескольких символов, за символьным классом следует указать *квантификатор*.

Квантификатор `{4}` повторяет `\d` четыре раза, как если бы мы использовали запись `\d\d\d\d` для совпадения с четыремя последовательными цифрами.

Во фрагменте `fullmatch` возвращает `None`, потому что `'986'` совпадает только с тремя последовательными цифровыми символами.

Квадратные скобки `[]` определяют *пользовательский символьный класс*, совпадающий с одним символом. Так, `[aeiou]` совпадает с гласной буквой нижнего регистра, `[A-Z]` — с буквой верхнего регистра, `[a-z]` — с буквой нижнего регистра и `[a-zA-Z]` — с любой буквой нижнего (верхнего) регистра.

Выполним простую проверку имени — последовательности букв без пробелов или знаков препинания. Проверим, что последовательность начинается с буквы верхнего регистра `( A–Z )`, а за ней следует *произвольное количество* букв нижнего регистра `( a–z )`:

````python
if re.fullmatch('[A-Z]*', 'Full HD 1920x1080'):
    print('yes')
else:
    print('no')

if re.fullmatch('[A-Z]*', 'FULLHD'):
    print('yes')
else:
    print('no')
````

Метасимволы в пользовательском символьном классе интерпретируются как литеральные символы, то есть как сами символы, не имеющие специального смысла.

Таким образом, символьный класс `[*+$]` совпадает с одним из символов `*` , `+` или `$`.

Квантификаторы `*` и `+` являются максимальными (*"жадными"*) — они совпадают с максимально возможным количеством символов.

Таким образом, регулярные выражения `[A-Z][a-z]+` совпадают с именами `'Al'` , `'Eva'` , `'Samantha'` , `'Benjamin'` и любыми другими словами, начинающимися с буквы верхнего регистра, за которой следует хотя бы одна буква нижнего регистра.

Квантификатор `?` совпадает **с нулем или одним вхождением** подвыражения.

Квантификатор `{n,}` совпадает *не менее чем* с `n` вхождениями подвыражения. Следующее регулярное выражение совпадает со строками, содержащими не менее трех цифр.

````python
if re.fullmatch(r'\d{3,}', 'Full HD 1920x1080'):
    print('yes')
else:
    print('no')
````

## Функция match

Эта функция ищет `pattern` в `string` и поддерживает настройки с помощью дополнительного `flags`.

Ниже можно увидеть синтаксис данной функции:

`re.match(pattern, string, flags=0)`

Функция `re.match` возвращает объект `match` при успешном завершении, или `None` при ошибке.

````python
re.match(r'FULL', 'FULL HD 1920x1080')
````

````python
re.match(r'FULL', 'FULL HD 1920x1080').group()
````

````python
result = re.match(r'FULL', 'FULL HD 1920x1080')

print(result.start())
print(result.end())
````

````python
re.match(r'FULL', 'HD FULL 1920x1080')
````

Возвращает объект соответствия, если ноль или более символов в **начале строки** подпадают под шаблон.

## Функция search

Эта функция выполняет поиск первого вхождения `pattern` внутри `string` с дополнительным `flags`.

Пример синтаксиса для этой функции:

`re.search(pattern, string, flags=0)`

Функция `re.search` возвращает объект `match` если совпадение найдено, и `None`, когда нет совпадений.

````python
re.search(r'FULL', 'FULL HD 1920x1080')
````

**`Python` предлагает две разные примитивные операции, основанные на регулярных выражениях: `match` выполняет поиск паттерна в начале строки, тогда как `search` выполняет поиск по всей строке.**

Метод search() ищет по всей строке, но возвращает только первое найденное совпадение.

````python
re.search(r'FULL', 'FULL HD 1920x1080 FULL')
````

````python
re.search(r'(\d+)\D(\d+)', 'IPS Panel Full HD / Touchscreen 1920x1080')
````

## Функция findall

Возвращает список всех найденных совпадений. У метода `findall()` нет ограничений на поиск в начале или конце строки.

Пример синтаксиса для этой функции:

`re.findall(pattern, string, flags=0)`

Функция `findall()` возвращает все неперекрывающиеся совпадения шаблона `pattern` в строке `string` в виде списка строк или список кортежей. Строка сканируется слева направо, и совпадения возвращаются в найденном порядке.

````python
re.findall(r'FULL', 'FULL HD 1920x1080 FULL')
````

````python
re.findall(r'[0-9]*[x][0-9]*', 'FULL HD 1920x1080 FULL')
````

````python
re.findall(r'[0-9]*[x][0-9]*', 'IPS Panel Touchscreen / 4K Ultra HD 3840x2160')
````

````python
re.findall(r'[0-9]*[x][0-9]*', '1920x1080 IPS Panel Touchscreen / 4K Ultra HD 3840x2160')
````

````python
def get_formatted_screen(text):
    result = re.findall(r'[0-9]*[x][0-9]*', text)
    return result[0]

df['ScreenResolution'].map(get_formatted_screen)
````

## Функция split

Этот метод разделяет строку по заданному шаблону.

Синтаксис метода:

`re.split(pattern, string, [maxsplit=0])`

````python
re.split(r'[\s]', '1920x1080 IPS Panel Touchscreen / 4K Ultra HD 3840x2160') # разделение по пробелу
````

````python
re.split(r'[\s]', '1920x1080 IPS Panel Touchscreen / 4K Ultra HD 3840x2160', maxsplit=1) # разделение по первому пробелу
````

````python
re.split(r'[\/]', '1920x1080 IPS Panel Touchscreen / 4K Ultra HD 3840x2160') # разделение по /
````

## Функция sub

Ищет шаблон в строке и заменяет его на указанную подстроку. Если шаблон не найден, строка остается неизменной.

Синтаксис:

`re.sub(pattern, repl, string)`

````python
df.head()
````

````python
re.sub(r'GB', '', '16GB')
````

````python
def get_formatted_Ram(value):
    value = re.sub(r'GB', '', value)
    return value

df['Ram'].map(get_formatted_Ram)
````

## Динамическое изменение типа данных столбцов

Если в DataFrame много столбцов и неизвестно, какие из них нужно преобразовать, можно использовать метод `apply()`, который применяет указанную функцию к каждому столбцу.

````python
df = df.apply(lambda col: col.astype(float) if col.str.replace('.','').str.isdigit().all() else col)
````

В этом примере мы применяем функцию, которая проверяет, содержат ли все строки в столбце только числовые символы, и если это так, преобразует столбец в числа с плавающей точкой.
